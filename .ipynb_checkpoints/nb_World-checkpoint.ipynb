```python
"""
covid-19 هذاتحليل لحالة
في تونس استناداً إلى البيانات المتاحة من جامعة جونز هوبكنز

MIT licensed by Ayoub Malek
"""
import warnings
import datetime
import matplotlib
import pandas as pd
import arabic_reshaper
import matplotlib.pyplot as plt
from bidi import algorithm as bidialg
from utils.covid_world import CovidWorld


# hide warnings
warnings.filterwarnings("ignore") 

# set style
matplotlib.rcParams.update({'font.size': 9})
plt.style.use('ggplot')
pd.set_option('expand_frame_repr', False)

# world 
world = CovidWorld()
world.parse_data()


# prints
print("DATA")
print(world.world_covid_df)

print("-------------------------")
print("WORST MORTALITY RECORDS")
print("سجلات الوفيات الأسوأ")
print("-------------------------")
print(world.world_covid_df.groupby(["country"]).max().sort_values(by = ['death_cases']).tail(10))

print("-------------------------")
print("WORST INFECTION RECORDS")
print("سجلات العدوى أسوأ")
print("-------------------------")
print(world.world_covid_df.groupby(["country"]).max().sort_values(by = ['confirmed_cases']).tail(10))
```

```python
# define list of dates
num_of_days = 7
base = datetime.datetime.today() - datetime.timedelta(days=num_of_days)
dates_list = [base + datetime.timedelta(days=x) for x in range(num_of_days)]


for date in dates_list:
    try:
        world.parse_data()
        world.plot_countries(filter_date=str(date.date()),
                             filter_countries=["Algeria", "Iran", "Sweden", "China", "Italy",
                                               "France", "Spain", "Tunisia", "Turkey",
                                               "Germany", "US", "United Kingdom"], 
                             title= bidialg.get_display(arabic_reshaper.reshape(u" فيروس كورونا حول العالم")))
    except Exception as e: 
        print(e)
```
